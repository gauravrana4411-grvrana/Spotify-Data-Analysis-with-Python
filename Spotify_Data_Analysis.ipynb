Import Required Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Importing the Dataset

```python
sp_tracks = pd.read_csv(r'D:\Project\Spotify Data Analysis\tracks.csv')
sp_artists = pd.read_csv(r'D:\Project\Spotify Data Analysis\artists.csv')
```

Identifying Null Values in the Dataset

```python
sp_tracks.info()
```

```python
#checking null in tracks data
sp_tracks.isnull().sum()
```

```python
# sp_tracks.dropna(subset='name',inplace=True)
sp_tracks['name']=sp_tracks['name'].fillna('Unknown')
sp_tracks.isnull().sum()
```

```python
#checking null in artists data
sp_artists.isnull().sum()
```

```python
sp_artists.info()
```

```python
sp_artists['followers']=sp_artists['followers'].fillna(sp_artists['followers'].mean())
sp_artists['name']=sp_artists['name'].fillna('Unknown')
sp_artists.isnull().sum()
```

Dataset Overview: Rows, Columns, Data Types, and Memory Usage

```python
#checking info in tracks data
sp_tracks.info()
```

```python
#checking info in artists data
sp_artists.info()
```

```python
sp_tracks['id_artists'] = (
    sp_tracks['id_artists']
    .astype(str)
    .str.replace(r"[\'\"\[\]]", "", regex=True)
    .str.strip()
)
```

Merging Two Dataset

```python
sp_merged = pd.merge(sp_tracks,sp_artists,how='inner',left_on='id_artists',right_on='id')

sp_merged
```

```python
sp_merged.info()
```

```python
sp_merged.rename(columns={'id_x': 'track_id','name_x':'track_name','popularity_x':'track_popularity',
                         'id_y':'artist_id','name_y':'artist_name','popularity_y':'artist_popularity'}, inplace=True)
```

```python
sp_merged.info()
```

```python
sp_merged['release_date'] = pd.to_datetime(
    sp_merged['release_date'],
    errors='coerce'
)

sp_merged['release_date'].info()
```

6. Visualization: Pearson Correlation Heatmap for Two Variables

```python
hm=sp_merged.select_dtypes(include='number').corr(method='pearson')
hm=round(hm,2)
plt.figure(figsize=(12,5))
sns.heatmap(hm, annot = True, cmap='coolwarm',linewidths=0.1,linecolor='black')
plt.title('Correlation Heatmap')
plt.show()
```

7. Creating a 4% Sample of the Entire Dataset

```python
sample_df=sp_merged.sample(int(0.004*len(sp_merged)))
print(len(sample_df))
sample_df['loudness']
```

8. Regression Plot of Loudness vs. Energy with Regression Line

```python
plt.figure(figsize=(8,4))
sns.regplot(data=sample_df, y='loudness', x='energy', color='#054907')
plt.title('Loudness vs Energy')
plt.xlabel('Energy')
plt.ylabel('Loudness (dB)')
plt.show()
```

9. Regression Plot of Popularity vs. Acousticness with Regression Line

```python
plt.figure(figsize=(8,4))
sns.regplot(data=sample_df, y='track_popularity', x='acousticness', color='#008000')
plt.title('Regression Plot - Popularity vs Acousticness Correlation')
plt.xlabel('Energy')
plt.ylabel('Loudness')
plt.show()
```

10. Adding a New Column to the Tracks Table

```python
sp_merged['years']=sp_merged['release_date'].dt.year.astype('Int64')
# sp_merged['years']=sp_merged['years'].astype(int)
sp_merged['years']
```

11. Graph: Number of Songs per Year

```python
# plt.figure(figsize=(12,8))
song_per_year_count=sp_merged.groupby('years')['track_name'].size().reset_index(name='count')
song_per_year_count.plot(x='years', y='count', kind='line', marker='o',color='purple')
plt.title('Songs Released Per Year')
plt.xlabel('Year')
plt.ylabel('Number of Songs')
plt.show()
```

```python
bins=[0,1900,1920,1940,1960,1980,2000,2020,2040]
labels=[1900,1920,1940,1960,1980,2000,2020,2040]
sp_merged['year_category']=pd.cut(sp_merged['years'],bins=bins,labels=labels)
sp_merged['year_category']
```

```python

song_per_year_category=sp_merged.groupby('year_category')['track_name'].size().reset_index(name='count')
song_per_year_category
```

```python
sns.barplot(data=song_per_year_category, x='year_category', y='count')
plt.xticks(rotation=90)
plt.title('Number of Songs Released Per Year Category')
plt.show()
```

```python
plt.figure(figsize=(15,5))
sns.countplot(data=sp_merged,x='years',palette='viridis')
plt.title('No of songs - per year')
plt.xlabel('Year')
plt.xlabel('Count')
plt.xticks(rotation=90)
plt.show()
```

12. Line Graph: Duration of Songs Over Each Year

```python
avg_duration_per_year =sp_merged.groupby('years')['duration'].mean().reset_index()

plt.figure(figsize=(15, 5))
sns.lineplot( data=avg_duration_per_year,x='years',y='duration',marker='o',color='purple')

plt.title('Average Song Duration per Year')
plt.xlabel('Year')
plt.ylabel('Duration (ms)')
plt.xticks(rotation=90)
plt.show()
```

```python
total_dr=sp_merged['duration']
plt.figure(figsize=(15,5))
sns.barplot(data=sp_merged,x='years',y='duration',palette='viridis',errorbar=None)
plt.title('Years vs Duration')
plt.xlabel('Years')
plt.ylabel('Duration')
plt.xticks(rotation=90)
plt.show()
```

13. Horizontal Bar Plot: Song Duration Across Different Genres

```python
genre_duration = sp_merged.groupby('genres')['duration'].sum().sort_values(ascending=False).reset_index()

top_genres = genre_duration.head(10)
```

```python
top_genres
```

```python
plt.figure(figsize=(10, 6))
# sns.set_palette('crest')

sns.barplot(data=top_genres, x='duration',y='genres',palette='viridis')

plt.title('Average Duration of Songs by Genre')
plt.xlabel('Duration')
plt.ylabel('Genre')
plt.tight_layout()
plt.show()
```

15. Bar Plot: Top Five Genres by Popularity

```python

Top = sp_merged.sort_values('track_popularity', ascending=False)[:5]

plt.figure(figsize=(12,6))
sns.barplot(data = Top,x = 'track_popularity',y = 'genres',palette='viridis')
plt.title('Genres by Popularity-Top 5')
plt.xlabel('Track Popularity')
plt.ylabel('Genres')
plt.show()
```
